```python
class Animal:
    def __init__(self):
        self.age = 1 

class Cat(Animal):
    def __init__(self):
        self.age = 3
        super().__init__()
        self.age += 1

    def show(self):
        print(self.age)

cat = Cat()
cat.show()
```

```python
class Animal:
    def __init__(self):
        self.name = "Calico"
        self._color = 'white'
        self._age = 1.5 

class Cat(Animal):
    def __init__(self):
        super().__init__()
        self._color  = 'black'
        self._age += 0.5

    def show(self):
        print(self.name)
        
        print(self._color)
        print(self._age)

cat = Cat()
cat.show()
```
