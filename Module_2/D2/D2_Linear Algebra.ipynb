```python
import cv2
import numpy as np

image = cv2.imread('FakeBackground.png', 1)
print(image.shape)

image = image.astype('float')

# process
image = image.mean(axis=2)

image = image.astype(dtype=np.uint8)
cv2.imwrite('result.png', image)

```

```python
import cv2
import numpy as np

image = cv2.imread('FakeBackground.png', 1)
print(image.shape)

image = image.astype('float')

# process
vector = np.array([1/3, 1/3, 1/3])      # lấy trung bình của 3 kênh màu
image = image@vector

image = image.astype(dtype=np.uint8)
cv2.imwrite('result2.png', image)

```

```python
import numpy as np

data = np.array([20, 180, 4, 70], dtype=np.uint8)

_, result = cv2.threshold(data, 100, 255, cv2.THRESH_BINARY)
print(result)

result[result == 0] = 1
print(result)
```
